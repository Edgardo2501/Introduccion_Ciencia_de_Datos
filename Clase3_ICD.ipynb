```python
# Cargar base de datos
import pandas as pd
df=pd.read_csv('https://raw.githubusercontent.com/jarsmp37/introds/main/Churn_Bank.csv')
df
```

Con el comando siguiente filtramos todos los valores diferentes de 'invalid_value'

```python
df2 = df[df["HasCrCard"]!='invalid_value']
df2
```

```python
df2['HasCrCard'].unique()
```

Con astype podemos modificar la columna de tipo 'object' a 'float'

```python
df2['HasCrCard']=df2['HasCrCard'].astype(float)
```

```python
df2 = df2.dropna(subset=["HasCrCard"])
df2
```

```python
df2['HasCrCard']=df2['HasCrCard'].astype(int)
```

```python
df2['HasCrCard'].unique()
```

Limpiamos la variable "CreditScore"

```python
df2 = df2[df2["CreditScore"]!='invalid_value']
df2
```

```python
df2['CreditScore'].unique()
```

```python
df2 = df2.dropna(subset=["CreditScore"])
df2
```

```python
df2['CreditScore']=df2['CreditScore'].astype(int)
```
