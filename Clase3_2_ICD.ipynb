```python
# Cargar base de datos
import pandas as pd
df=pd.read_csv('https://raw.githubusercontent.com/jarsmp37/introds/main/Churn_Bank.csv')
df
```

```python
# Eliminamos una fila mediante el indexado
# Elimina filas con índices 0 y 1

df = df.drop([0,1])
df = df.drop([12203])
df
```

```python
# Con reset_index reestablecemos el índice
df = df.reset_index(drop=True)
df
```

```python
# Eliminar columnas
df2 = df.drop(columns=['RowNumber'])
df2
```

```python
# Renombrar columnas
df2 = df2.rename(columns={'Surname':'Apellido'})
df2
```

```python
# Renommbrar multiples columnas
df2 = df2.rename(columns={'Geography':'Pais','Gender':'Genero','Age':'Edad'})
df2
```

```python
# Crear un rango de fechas
start_date = '2024-08-01'
end_date = '2025-08-01'
date_range = pd.date_range(start=start_date, end=end_date, periods=12202)

# Convertir las fechas a formato string
date_range_str = date_range.strftime('%Y-%m-%d')
date_range_str
```

```python
# Anexar una columna llamada Fecha
df2['Fecha']=date_range_str
df2.head()
```

```python
df2.info()
```

```python
# Le damos formato date a la columna nueva 'Fecha'
df2['Fecha']=pd.to_datetime(df2['Fecha'])
df2.info()
```

```python
# Filtrar una fecha específica
df_filtrado = df2[df2['Fecha'] >= '2024-12-01']
df_filtrado
```

```python
# Filtrar un rango específico de fecha
df_rangofecha = df2[(df2['Fecha'] >= '2024-10-01') & (df2['Fecha'] <= '2025-03-30')]
df_rangofecha
```

```python
# Extraer el año
df2['año'] = df2['Fecha'].dt.year

# Extraer el mes
df2['mes'] = df2['Fecha'].dt.month

# Extraer el día
df2['dia'] = df2['Fecha'].dt.day

# Extraer el día de la semana
df2['dianombre'] = df2['Fecha'].dt.day_name()

df2.head()
```

```python
# Calcular la diferencia en días entre dos fechas

df2['days_difference'] = (df2['Fecha'] - pd.Timestamp('2024-01-01')).dt.days
df2.tail()
```

```python
# Diccionario para traducir los nombres de los días de la semana
traduccion = {
    'Monday':'Lunes',
    'Tuesday': 'Martes',
    'Wednesday': 'Miercoles',
    'Thursday': 'Jueves',
    'Friday': 'Viernes',
    'Saturday': 'Sabado',
    'Sunday': 'Domingo'
}
```

```python
# Reemplazar los nombres de los días en la columna 'dianombre'
df2['dianombre'] = df2['dianombre'].replace(traduccion)
```

```python
# Traducir los nombres del genero.

traduccion2 = {
    'Female': 'Femenino',
    'Male': 'Masculino'
}
```

```python
# Reemplazar los nombres del género en la columna 'Genero'
df2['Genero'] = df2['Genero'].replace(traduccion2)
```

```python
# Traducir los nombres de France, Spain y Germany a español.
traduccion3 = {
    'France': 'Francia',
    'Spain': 'España',
    'Germany': 'Alemania'
}
```

```python
# Reemplazar los nombres del género en la columna 'Genero'
df2['Pais'] = df2['Pais'].replace(traduccion3)
df2
```

```python
# Guardar los resultados en un csv
df2.to_csv("Base_limpia.csv",index=False)
```
